```python
# convert original problem into 81x9 matrix
def convert(df):
    import numpy as np
    import pandas as pd
    columns = [i for i in range(1,10)]
    rows = [i for i in range(1,82)]
    temp = pd.DataFrame(np.zeros((81,9)),index = rows, columns = columns)
    for i in range(1,10):
        temp.loc[1+9*(i-1):i*9] = np.where(df==i,1,0)
    return temp
df_new = convert(df)
df_new
```
