```python
# import basic package
import numpy as np
import pandas as pd

# import package for image recognition
from sklearn.neighbors import KNeighborsClassifier
import cv2
from PIL import Image
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import fetch_mldata

# import package for solving optimization
from pulp import *

# import package for GUI
from tkinter import *
import tkinter.filedialog


###############code start#################
#Prepare KNN model for Image recognition
mnist = fetch_mldata('MNIST original', data_home='C:/Users/DELL/Desktop/Python_tools') #change to your own directory
X = mnist.data
X = np.array(X)
X.reshape(70000,784)
y = mnist.target
y = np.array(y)
y.reshape(70000,1)

knn = KNeighborsClassifier(n_neighbors=30,weights='distance')
knn.fit(X,y)
print('KNN is Ready!')


#Prepare the GUI
m1 = []
m2 = []
I1 = False
I2 = False

#The main funcition for the whole program
def main():
    key = DisplayMainwindowAndGetInput()
    try: 
        if key == '0426':
            n = 9
            newwindow(n)
            if I1 == True:
                mlist = convert(n)
                displayTheAnswer(n, mlist)
            if I2 == True:
                displayTheAnswer2(n, m2)
                mlist = convert(n)
                displayTheAnswer(n, mlist)  
        else:
            print('Wrong Keys')
    except:
        print("some things is wrong, please contact tech supporter by yc3526@columbia.edu")
            
#The welcome page and fetch the passward            
def DisplayMainwindowAndGetInput():
    def pp():
        window.destroy()
    window = Tk()
    window.title('Sudoku Solver by Python Group')
    ntext = StringVar()
    lb1 = Label(window, text="""Welcome to the most Powerful Sudoku Solver in the world!""")
    lb1.grid(row=0, column=1, padx=5, pady=5)
    lb2 = Label(window, text="""Please enter your passwords to get full access to the services!""")
    lb2.grid(row=1, column=1, padx=5, pady=5)
    ent = Entry(window, width=15, textvariable=ntext)
    ent.grid(row=2, column=1, padx=5, pady=5)
    bnt = Button(window, text='Enter', command=pp)
    bnt.grid(row=3, column=0, columnspan=2, pady=10)
    window.mainloop()
    return ntext.get()

# The image recogtion program for a sepecific number
def knn_predict(filename):
    # resize image and export image
    pic = cv2.imread(filename)
    pic = cv2.resize(pic, (28, 28), interpolation=cv2.INTER_CUBIC)
    cv2.imwrite('im.png',pic)
    
    from PIL import Image
    # reopen image
    img = Image.open("im.png")
    
    # gray
    img = img.convert("L")
    
    # convert to matrix
    I = np.asarray(img)
    
    # flatten
    raw = I.reshape(1,784).tolist()[0]
    a = sum(raw)
    raw = np.array(raw).reshape(1,784)
    
    if a <= 255*10:
        return 0
    
    # use KNN model
    result_knn = knn.predict(raw)
    
    return int(result_knn[0])    

# The image recogtion program for the entire graph
def Image_Recognition(directory):
    
    from PIL import Image
    # read the original image from the directory
    image_sudoku_original = cv2.imread(directory)
    
    # grey the image
    image_sudoku_gray = cv2.cvtColor(image_sudoku_original,cv2.COLOR_BGR2GRAY)
    I = cv2.adaptiveThreshold(image_sudoku_gray,255,1,1,11,15)
    
    # convert the image into dataframe
    df = pd.DataFrame(I)
    df = df/255
    df = df.astype(int)
    
    # get rid of the outer frame
    columns = df.sum()
    start_col = 0
    end_col = 0
    
    for i in range(len(columns)): # if sum of 'cubics' is larger than half of the length, then it is a line
        if columns.iloc[i] > (len(columns)/2) and columns.iloc[i+1] < (len(columns)/2):
            break
    start_col = i+1
    for j in range(len(columns)-1,1,-1):
        if columns.iloc[j] > (len(columns)/2) and columns.iloc[j-1] < (len(columns)/2):
            break
    end_col = j
    
    
    rows = df.sum(1)
    start_row = 0
    end_row = 0
    
    for s in range(len(rows)):
        if rows.iloc[s] > (len(rows)/2) and rows.iloc[s+1] < (len(rows)/2):
            break
    start_row = s+1
    
    for q in range(len(rows)-1,1,-1):
        if rows.iloc[q] > (len(rows)/2) and rows.iloc[q-1] < (len(rows)/2):
            break
    end_row = q-1
    
    df1 = df.iloc[start_row:end_row,start_col:end_col] # the new df without outer frame
    
    # reindex df1 (let it start with 0)
    df1.index = range(len(df1))
    df1.columns = range(len(df1.columns))
    
    # delete the inner frames for rows
    df1_rows = pd.DataFrame(np.array(df1.sum(1))) 
    df1_rows_frame = df1_rows[df1_rows[0] > len(df1.sum(1))/2]
    rows_to_delete = np.array(df1_rows_frame.index)
    df2 = df1.drop(df1_rows.index[rows_to_delete])
    
    # delete the inner frames for colums
    df1_columns = pd.DataFrame(np.array(df1.sum())) 
    df1_columns_frame = df1_columns[df1_columns[0] > len(df1.sum())/2]
    columns_to_delete = np.array(df1_columns_frame.index)
    df3 = df2.drop(df1_columns.index[columns_to_delete],axis = 1) 
    df3 = df3.reset_index(drop=True)
    df3.columns = range(df3.shape[1])
    
    # segment out the indiviual numbers
    c = len(df3.sum())
    r = len(df3.sum(1))
    predict_list = list()
    for i in range(9):
        for j in range(9):
            segments = df3.iloc[i*r//9 : (i+1)*r//9, j*c//9 : (j+1)*c//9]
            each_seg = cv2.imwrite('each_seg.png',segments.values*255)
            predict_num = knn_predict('each_seg.png')
            predict_list.append(predict_num)
    
    return pd.DataFrame(np.array(predict_list).reshape(9,9))


# The GUI and the integer programming for solving Sudoku Problem
def newwindow(n):
    newwindow = Tk()
    newwindow.title('Your Choice!')
    def pp1():  
        newwindow.destroy() 
        global I1
        I1 = True
    global m1

    for i in range(n):
        line = []
        for j in range(n):
            temporarytext1 = StringVar()
            ent = Entry(newwindow, width=3, textvariable=temporarytext1)
            ent.grid(row=i, column=j, padx=5, pady=5)
            line.append(temporarytext1)
        m1.append(line)
    btn = Button(newwindow, text='where amazing happen!', command=pp1)
    btn.grid(row=int(n/2)-1, column=n, padx=5)
    
    
    def load():
        try:
            filename = tkinter.filedialog.askopenfilename()
            print(filename)
            
            global m2
            m2 = []
            dataf = Image_Recognition(filename)
            m2 = dataf.values.tolist()
            print('KNN prediction finished!')
            newwindow.destroy()
            global I2
            I2 = True
            
        except:
            print('Error')
            None

    btn2 = Button(newwindow, text='load your image!', command=load)
    btn2.grid(row=int(n/2)+1, column=n, padx=5) 
    newwindow.mainloop()


def convert(n):
    global m1
    
    def clean(x):
        try:
            a = int(x)
            return a
        except:
            return 0  
    mm1 = [[clean(j.get()) for j in i] for i in m1]
    columns = [i for i in range(1,10)]
    rows = [i for i in range(1,10)]
    
    df = pd.DataFrame(mm1,index=rows,columns=columns)

    def final_convert(df):

        # convert df(9x9) into dt_new(81x9)
        def convert(df):
            columns = [i for i in range(1,10)]
            rows = [i for i in range(1,82)]
            temp = pd.DataFrame(np.zeros((81,9)),index = rows, columns = columns)
            for i in range(1,10):
                temp.loc[1+9*(i-1):i*9] = np.where(df==i,1,0)
            return temp

        # create function make sure each point has only one value
        def clean_each(df_new):
            #Create A
            columns = [i for i in range(1,730)]
            rows = [i for i in range(1,82)]
            A = pd.DataFrame(np.zeros((81,729)),index = rows, columns = columns)

            #Create b
            columns = [1]
            rows = [i for i in range(1,82)]
            b = pd.DataFrame(np.zeros((81,1)),index = rows, columns = columns)  

            for i in range(1,10): #columns
                for j in range(1,10): #rows
                    result = []
                    for m in range(1,10):
                        result.append(df_new.loc[j+9*(m-1),i])
                        col = 9*(j-1)+81*(m-1)+i
                        A.loc[9*(i-1)+j,col] = 1
                        b.loc[9*(i-1)+j] = 1                         
                    if sum(result) == 0:
                        continue
                    elif sum(result) != 1:
                        print('Error')
                    else:      
                        index = result.index(1)+1
                        col = 9*(j-1)+81*(index-1)+i
                        A.loc[9*(i-1)+j,col] = 0
                        b.loc[9*(i-1)+j] = 0
            return A,b

        # functions make sure each row has only one value
        def clean_row(df_new):
        #create row_check R
            columns = [i for i in range(1,730)]
            rows = [i for i in range(1,82)]
            R = pd.DataFrame(np.zeros((81,729)),index = rows, columns = columns)
        #creat row_check b    
            columns = [1]
            rows = [i for i in range(1,82)]
            b = pd.DataFrame(np.ones((81,1)),index = rows, columns = columns) 
        #check
            for i in range(1,82):
                result = []
                for j in range(1,10):
                    result.append(df_new.loc[i,j])
                    R.loc[i,9*(i-1)+j] = 1
                if sum(result) == 1:
                    index = result.index(1)+1
                    R.loc[i,9*(i-1)+index] = 0
                    b.loc[i] = 0
            return R,b

        # functions make sure each column has only one value
        def clean_column(df):
        #create row_check A
            columns = [i for i in range(1,730)]
            rows = [i for i in range(1,82)]
            C = pd.DataFrame(np.zeros((81,729)),index = rows, columns = columns)
        #creat row_check b    
            columns = [1]
            rows = [i for i in range(1,82)]
            b = pd.DataFrame(np.ones((81,1)),index = rows, columns = columns) 
        #check
            for i in range(0,9):
                for j in range(1,10):
                    result = []
                    for w in range(0,9):
                        result.append(df[j][i*9+1+w])
                        C.loc[i*9+j,j+81*i+9*w] = 1
                        if sum(result) ==1:
                            index = result.index(1)+1
                            C.loc[i*9+j,j+81*i+9*(index-1)]=0
                            b.loc[i*9+j]=0
            return C,b

        # functions make sure the sum of each 3 by 3 grid equals to 1
        def clean_grid(df_new):
            #rearrange the index of the dataset df_new
            df_new.index = [i for i in range(81)]
            df_new.columns = [i for i in range(9)]
        #create an empty matrix G for grid check
            columns = [i for i in range(729)]
            rows = [i for i in range(81)]
            G = pd.DataFrame(np.zeros((81,729)),index = rows, columns = columns)
        #create an empty vertical array b 
            columns = [1]
            rows = [i for i in range(81)]
            b = pd.DataFrame(np.ones((81,1)),index = rows, columns = columns) 
        #check
            for i in range(81):#ith grid, and i is also the number of row of G and the number of constraint
                #The coordinates of the first element in the grid:
                col = (i % 3) *3
                row = (i // 3) * 3
                #now we creat a list containing the elements of the grid in df_new
                grid_list =list()
                for j in range(3):
                    for k in range(3):
                        row_num = row + j
                        col_num = col + k
                        grid_list.append(df_new.loc[row_num, col_num])
                        # Also we let the coefficient of the corresponding item in G to be 1
                        G.loc[i,row_num * 9 + col_num] = 1
                #now check if all the elements in grid_list are zeros
                sum_list = sum(grid_list)
                if sum_list == 0:
                    b.loc[i,1] = 1
                    continue
                elif sum_list == 1:
                    b.loc[i,1] = 0
                    #we find out which element in sum_list equal to 1 and let corresponding coefficient in G to be 0
                    index_one = grid_list.index(1)
                    row_one = row + index_one // 3
                    col_one = col + index_one % 3
                    G.loc[i,row_one * 9 + col_one] = 0
            #To be compatible with the function defined above, we change the index of G and df_new back
            df_new.index = [i for i in range(1,82)]
            df_new.columns = [i for i in range(1,10)]
            G.index = [i for i in range(1,82)]
            G.columns = [i for i in range(1,730)]
            b.index = [i for i in range(1,82)]
            return G,b

        def convert_back(df_new):
            columns = [i for i in range(1,10)]
            temp = pd.DataFrame(np.zeros((9,9)),columns = columns)
            for i in range(1,10):
                temp = temp + i*df_new.loc[1+(i-1)*9:i*9].reset_index(drop=True)
            temp.index = [i for i in range(1,10)]
            return temp.astype(int)


        df_new = convert(df)
        A1, b1 = clean_each(df_new)
        A2, b2 = clean_row(df_new)
        A3, b3 = clean_column(df_new)
        A4, b4 = clean_grid(df_new)

        A1 = A1.values;A2 = A2.values;A3 = A3.values;A4 = A4.values
        A1_list = A1.tolist();A2_list = A2.tolist();A3_list = A3.tolist();A4_list =A4.tolist()
        A1_list.extend(A2_list);A1_list.extend(A3_list);A1_list.extend(A4_list)
        A = A1_list
        columns = [i for i in range(1,730)]
        rows = [i for i in range(1,325)]
        A = pd.DataFrame(A,index = rows, columns = columns)

        b1_list = b1.values.tolist();b2_list = b2.values.tolist();b3_list = b3.values.tolist();b4_list = b4.values.tolist()
        b1_list.extend(b2_list);b1_list.extend(b3_list);b1_list.extend(b4_list)
        b = b1_list
        columns = [1]
        rows = [i for i in range(1,325)]
        b = pd.DataFrame(b,index = rows, columns = columns)

        A_list = A.values.tolist()
        b = b.values.tolist()
        FOOTNOTES = list(range(729))
        prob = LpProblem("SudokuSolver", LpMaximize)
        x = LpVariable.dicts("ZeroOneVariables", FOOTNOTES, 0, 1, LpBinary)
        prob += lpSum(x[i] for i in FOOTNOTES)
        for i in range(324):
            prob += lpSum(A_list[i][j] * x[j] for j in FOOTNOTES) == b[i][0]
        prob.solve()
        print(pulp.LpStatus[prob.status])
        if pulp.LpStatus[prob.status] == 'Optimal':
            df_result_list = list()
            for i in range(81):# i is the row of the resulted dataframe
                row_list= list()
                for j in range(9): # j is the column of the resulted dataframe
                    row_list.append(x[i * 9 + j].varValue)
                df_result_list.append(row_list)
            len(df_result_list)
            #Now we convert df_result_list into DataFrame
            rows = [i for i in range(1, 82)]
            columns = [i for i in range(1, 10)]
            df_result = pd.DataFrame(df_result_list, index = rows, columns = columns)
            df_final = convert_back(df_result)
            return df_final
    df_final = final_convert(df)
    output = df_final.values.tolist()

    return output

# The GUI for displaying result for hand-inport sudoku problem
def displayTheAnswer(n, mlist):
    nnwindow = Tk()
    if mlist == None:
        nnwindow.title('No solution! I am 100% sure!')
    else:
        nnwindow.title('Solved!')
        for i in range(n):
            for j in range(n):
                temtext = StringVar()
                ent = Entry(nnwindow, width=3, textvariable=temtext ,state = 'readonly')
                ent.grid(row=i, column=j, padx=5, pady=5)
                temtext.set(mlist[i][j])
    nnwindow.mainloop()

# The GUI for displaying result for image-import sudoku problem
def displayTheAnswer2(n, mlist):
    nnnwindow = Tk()
    def pp1():
        nnnwindow.destroy()
        
    def clean2(t):
        if t == 0:
            return ''
        return t   
    
    btn = Button(nnnwindow, text='where amazing happen!', command=pp1)
    btn.grid(row=int(n/2)-1, column=n, padx=5)
    
    global m1
    m1 = []
    if mlist == None:
        nnnwindow.title('Sorry! Recognition Failed!')
    else:
        nnnwindow.title('I got YOU!')
        for i in range(n):
            line = []
            for j in range(n):
                temtext = StringVar()
                ent = Entry(nnnwindow, width=3, textvariable=temtext)
                ent.grid(row=i, column=j, padx=5, pady=5)
                temtext.set(clean2(mlist[i][j]))
                line.append(temtext)
            m1.append(line)
    
    
    nnnwindow.mainloop()


main()
```
